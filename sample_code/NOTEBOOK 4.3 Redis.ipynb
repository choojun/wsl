```python
import redis
# Create a connection to the Redis server
r = redis.Redis(host='localhost', port=6379, db=0)
```

```python
# Set a key-value pair
r.set('mykey', 'myvalue')
# Retrieve the value of the key
value = r.get('mykey')
print(value.decode('utf-8'))  # Outputs: myvalue
```

```python
# Set multiple key-value pairs
r.mset({"key1": "value1", "key2": "value2"})
# Get multiple values using their keys
values = r.mget(["key1", "key2"])
for val in values:
   print(val.decode('utf-8'))
```

```python
# Delete a key-value pair
r.delete('mykey')
# Check if a key exists
exists = r.exists('mykey')
print(exists)  # Outputs: 0
```

```python
# You can replace this later to split based on punctuations, etc.
def tokenise(text):
    return text.split()
```

```python
def return_full_form(word):
    key = str(word).lower()
    if r.exists(key):
        return r.get(key).decode('utf-8')
    return word
```

```python
def replace_short_forms(text):
    tokens = tokenise(text)
    result_list = [return_full_form(t) for t in tokens]
    print(f"***TRACE: result_list: {result_list}")
    print(f"***TRACE: type(result_list): {type(result_list)}")
    print(f"***TRACE: result_list[0]: {result_list[0]}")
    print(f"***TRACE: type(result_list[0]): {type(result_list[0])}")
    return " ".join(result_list)
```
